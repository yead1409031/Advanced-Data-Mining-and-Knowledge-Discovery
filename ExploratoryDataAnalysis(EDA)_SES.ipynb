**Data Filtering - Simple Exponential Smoothing**

```python
def exp_smooth(alpha):
    ft = 0
    forecast = []
    for i in range(0,len(data1)):
        if(ft == 0):
            ft = data1['rssi'][i]
            forecast.append(ft)
        else:
            ft = alpha*(data1['rssi'][i]) + (1-alpha)*(ft)
            forecast.append(ft)
    return forecast
data1['ses_rssi'] = exp_smooth(0.2)  #### Try to change the smoothing parameter. It should be range between 0 to 1 ####
print(data1)



def exp_smooth(alpha):
    ft = 0
    forecast = []
    for i in range(0,len(data2)):
        if(ft == 0):
            ft = data2['rssi'][i]
            forecast.append(ft)
        else:
            ft = alpha*(data2['rssi'][i]) + (1-alpha)*(ft)
            forecast.append(ft)
    return forecast
data2['ses_rssi'] = exp_smooth(0.2)
print(data2)


def exp_smooth(alpha):
    ft = 0
    forecast = []
    for i in range(0,len(data3)):
        if(ft == 0):
            ft = data3['rssi'][i]
            forecast.append(ft)
        else:
            ft = alpha*(data3['rssi'][i]) + (1-alpha)*(ft)
            forecast.append(ft)
    return forecast
data3['ses_rssi'] = exp_smooth(0.2)
print(data3)



def exp_smooth(alpha):
    ft = 0
    forecast = []
    for i in range(0,len(data4)):
        if(ft == 0):
            ft = data4['rssi'][i]
            forecast.append(ft)
        else:
            ft = alpha*(data4['rssi'][i]) + (1-alpha)*(ft)
            forecast.append(ft)
    return forecast
data4['ses_rssi'] = exp_smooth(0.2)
print(data4)
```

**Data Transformation using Path Loss Model**

```python
def pathlossmodel(rssi_value, n, c):
    """
    Calculates the estimated distance based on an RSSI value using the pathloss model.

    Args:
        rssi_value: A single RSSI value.
        n: Pathloss exponent.
        c: Reference constant.

    Returns:
        The estimated distance.
    """
    estimated_distance = 10 ** ((rssi_value - c) / (-10 * n))
    return estimated_distance


# Define pathloss parameters
n = 0.8891
c = -64.63

################## Zone A ###################################
# updated_data1  containing rolled features and 'zone' column
# Get only the rolled features
rssi_features1 = data1.drop(columns=['zone'])

# Apply pathlossmodel to rolled features
rssi_features1_distance1 = rssi_features1.apply(lambda x: pathlossmodel(x, n, c))

# Combine with the 'zone' column
data1_dist = pd.concat([data1['zone'], rssi_features1_distance1], axis=1)

# Rename 'rssi' column as 'dist'
data1_dist.rename(columns={'ses_rssi': 'ses_dist'}, inplace=True)

# Print the result DataFrame
print(data1_dist)
data1_dist.to_csv('data1_distoutput.csv', index=False)


################## Zone B ###################################
# updated_data1  containing rolled features and 'zone' column
# Get only the rolled features
rssi_features2 = data2.drop(columns=['zone'])

# Apply pathlossmodel to rolled features
rssi_features2_distance2 = rssi_features2.apply(lambda x: pathlossmodel(x, n, c))

# Combine with the 'zone' column
data2_dist = pd.concat([data2['zone'], rssi_features2_distance2], axis=1)

# Rename 'rssi' column as 'dist'
data2_dist.rename(columns={'ses_rssi': 'ses_dist'}, inplace=True)

# Print the result DataFrame
print(data2_dist)

################## Zone C ###################################
# updated_data1  containing rolled features and 'zone' column
# Get only the rolled features
rssi_features3 = data3.drop(columns=['zone'])

# Apply pathlossmodel to rolled features
rssi_features3_distance3 = rssi_features3.apply(lambda x: pathlossmodel(x, n, c))

# Combine with the 'zone' column
data3_dist = pd.concat([data3['zone'], rssi_features3_distance3], axis=1)

# Rename 'rssi' column as 'dist'
data3_dist.rename(columns={'ses_rssi': 'ses_dist'}, inplace=True)

# Print the result DataFrame
print(data3_dist)

################## Zone D ###################################
# updated_data1  containing rolled features and 'zone' column
# Get only the rolled features
rssi_features4 = data4.drop(columns=['zone'])

# Apply pathlossmodel to rolled features
rssi_features4_distance4 = rssi_features4.apply(lambda x: pathlossmodel(x, n, c))

# Combine with the 'zone' column
data4_dist = pd.concat([data4['zone'], rssi_features4_distance4], axis=1)

# Rename 'rssi' column as 'dist'
data4_dist.rename(columns={'ses_rssi': 'ses_dist'}, inplace=True)

# Print the result DataFrame
print(data4_dist)
```

**Merge four datasets**

```python
###### Merge all datasets ########
raw_distance = pd.concat([data1_dist, data2_dist, data3_dist ,data4_dist], ignore_index=True)
print(raw_distance)
```

**Machine Learning Models**

1.   Decision Tree
2.   Naive Bayes
3.   Logistic Regression

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import seaborn as sns
import matplotlib.pyplot as plt

# Split data into features (X) and target (y)
X = raw_distance['ses_dist']
y = raw_distance['zone']


# Split the dataset into training and testing sets (70:30 ratio)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Reshape X_train and X_test
X_train = X_train.values.reshape(-1, 1)
X_test = X_test.values.reshape(-1, 1)

# Train Decision Tree model
dt_model = DecisionTreeClassifier(criterion='gini')
dt_model.fit(X_train, y_train)
dt_pred = dt_model.predict(X_test)

# Train Naive Bayes model
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
nb_pred = nb_model.predict(X_test)

# Train the logistic regression model
logreg_model = LogisticRegression(multi_class='ovr', max_iter=1000)  # Use one-vs-rest (OvR)
logreg_model.fit(X_train, y_train)
logreg_pred = logreg_model.predict(X_test)


# Evaluate models
models = {'Decision Tree': dt_pred, 'Naive Bayes': nb_pred,'Logistic Regression':logreg_pred}
for model_name, y_pred in models.items():
    print(f"Model: {model_name}")
    cm = confusion_matrix(y_test, y_pred)
    #print(cm)
    print(confusion_matrix(y_test, y_pred))
    print("Accuracy:", "{:.2%}".format(accuracy_score(y_test, y_pred)))
    print("Precision:", "{:.2%}".format(precision_score(y_test, y_pred, average='weighted',zero_division=1)))
    print("Recall:", "{:.2%}".format(recall_score(y_test, y_pred, average='weighted')))
    print("F1 Score:", "{:.2%}".format(f1_score(y_test, y_pred, average='weighted')))
    print("\n")
    # Get unique class labels
    unique_labels = sorted(y_test.unique())

    # Visualize confusion matrix as a heatmap with class labels
    plt.figure(figsize=(8, 5))
    sns.heatmap(cm, annot=True, cmap='Blues', fmt='g', cbar=False, xticklabels=unique_labels, yticklabels=unique_labels)
    plt.title(f'Confusion Matrix for {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    #plt.ylabel('Actual')
    #plt.show()
```
